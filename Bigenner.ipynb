# What is Python?

**Python** is a high-level, interpreted programming language known for its simple and readable syntax. It emphasizes code readability and reduces development time compared to languages like C++ or Java.

## Key Characteristics:
- **Easy to learn** — Clean, intuitive syntax that reads like English
- **Interpreted language** — Code runs line-by-line without compilation
- **Versatile** — Used for web development, data science, AI/ML, automation, and more
- **Dynamically typed** — You don't need to declare variable types explicitly
- **Object-oriented** — Supports classes and inheritance

# Python Developer/Creator

**Guido van Rossum** is the creator and original developer of Python.

## About Guido van Rossum:

- **Born:** 1956 in Netherlands
- **Education:** University of Amsterdam (mathematics and computer science)
- **Created Python:** In 1989 while working at CWI (Centrum Wiskunde & Informatica) in Amsterdam
- **Motivation:** He wanted a language with clear syntax that was easy to learn and use
- **First Release:** Python 0.9.0 in 1991
- **Legacy:** Nicknamed the "Benevolent Dictator For Life" (BDFL) due to his leadership in Python's development for decades

## Why He Created Python:

Guido designed Python as a successor to the **ABC programming language**, aiming to create something:
- More flexible and powerful
- Easier to learn and read
- Suitable for both beginners and experienced programmers

## Current Status:

- Guido stepped back from BDFL role in 2019 but remained involved
- Python is now managed by the Python Software Foundation (PSF)
- The language continues to evolve with contributions from thousands of developers worldwide

So in short: **Guido van Rossum** developed Python in 1989! 🐍

**Strings**
 In Python, a string is a sequence of characters enclosed in quotes.
 You can use single quotes, double quotes, or triple quotes for multi-line strings.

 Examples:

 text = "Hello"
 
 text2 = 'World'
 
 multiline = """Line 1
 
 Line 2"""
 
 
 Strings are immutable, so you cannot change a character in place.
 Common operations:
 - concatenation: "Hello" + " " + "World"
 - repetition: "ha" * 3
 - slicing: s[0:5], s[-3:]
 - methods: .upper(), .lower(), .strip(), .replace(), .split()


```python
message = "Hello\', world!"
print(message)
```

```python
message = """Hello, world
This is a multi-line string.
It can span multiple lines without needing special characters."""
print(message)
```
